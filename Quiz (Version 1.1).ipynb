The question are not randomly generated, but when you failed at any point(level) it is GAME OVER.
When you failed a question at any point the game restart automatically, also you can now shutdown the GAME at any point.

```python
Score = None
def quiz1 ():
    print("\nQuestions 1")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    
    Ans1 = input("Enter your Answer (i.e A/B/C/D) or 'quit' to Shutdown the Quiz")
    shutDown = None
    match Ans1:
        case "A":
            print("Correct")
            Score = 100
            print (f"Score: {Score}pts    Level: 1")
        case "B":
            print("Incorrect")
            Score = 0
            print (f"Score: {Score}pts    Level: 1")
        case "C":
            print("Incorrect")
            Score = 0
            print (f"Score: {Score}pts    Level: 1")
        case "D":
            print("Incorrect")
            Score = 0
            print (f"Score: {Score}pts    Level: 1")
        case "quit":
            Score = 0
            shutDown = "q"
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz1 ()
    return Score, shutDown

def quiz2 (Score):
    print("\nQuestions 2")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    
    Ans2 = input("Enter your Answer (i.e A/B/C/D)")
    shutDown = None   
    match Ans2:
        case "A":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 2")
        case "B":
            print("Correct")
            Score += 100
            print (f"Score: {Score}pts    Level: 2")
        case "C":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 2")
        case "D":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 2")
        case "quit":
            Score += 0
            shutDown = "q"
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz2 (Score)
    return Score, shutDown

def quiz3 (Score):
    print("\nQuestions 3")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    Ans3 = input("Enter your Answer (i.e A/B/C/D)")
    
    shutDown = None    
    match Ans3:
        case "A":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "B":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "C":
            print("Correct")
            Score += 200
            print (f"Score: {Score}pts    Level: 3")
        case "D":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "quit":
            Score += 0
            shutDown = "q"
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz3 (Score)
    return Score, shutDown

def quiz4 (Score):
    print("\nQuestions 4")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    Ans3 = input("Enter your Answer (i.e A/B/C/D)")

    shutDown = None
    match Ans3:
        case "A":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "B":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "C":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "D":
            print("Correct")
            Score += 400
            print (f"Score: {Score}pts    Level: 3")
        case "quit":
            Score += 0
            shutDown = "q"
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz3 (Score)
    return Score, shutDown

def quizGame():
    Score = quiz1()
    
    if Score[1] == "q":
        final = Score[0]
        SD = "Shutdown"
    
    else:
        while Score[0] == 0:
            print("\nGame Over ReStarting.........................")
            Score = quiz1()
            if Score[1] == "q":
                break
    
    if Score[1] == "q":
        final = Score[0]
        SD = "Shutdown"
    
    else:     
        Score2 = quiz2(Score[0])
        
        if Score2[1] == "q":
            final = Score[0]
            SD = "Shutdown"
        
        else:
            if Score2[0] == 100:
                print("\nGame Over ReStarting.........................")
                final = quizGame()
                
                if final[1] == "Shutdown":
                    SD = "Shutdown" 
                else:
                    SD = None
                final = final[0]
            
            else:
                Score3 = quiz3(Score2[0])
                
                if Score3[1] == "q":
                    final = Score2[0]
                    SD = "Shutdown"
                
                else:
                    if Score3[0] == 200:
                        print("\nGame Over ReStarting.........................")
                        final = quizGame()
                        
                        if final[1] == "Shutdown":
                            SD = "Shutdown" 
                        else:
                            SD = None
                        final = final[0]
                        
                    else:
                        final = quiz4(Score3[0])
                        
                        if final[1] == "q":
                            final = Score3[0]
                            SD = "Shutdown"  
                        
                        else:
                            if final[0] == 400:
                                print("\nGame Over ReStarting.........................")
                                Score4 = quizGame()
                                final = Score4[0]
                            
                            else:
                                SD = None
                                
                                print("\nYour Score BreakDown")
                                print(f"Level 1: {Score[0]}")
                                print(f"Level 2: {Score2[0]}")
                                print(f"Level 3: {Score3[0]}")
                                print(f"Level 4: {final[0]}")
                                
                                final = final[0]
        
    return final, SD
    

totalScore = quizGame()

if totalScore[1] == "Shutdown":
    print("\nShutting Down...................... GoodBye!!!")
    print(f"Total Point: {totalScore[0]}")
    

else:
    print(f"Total Point: {totalScore[0]}")
```
