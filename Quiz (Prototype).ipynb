The question are not randomly generated, but when you failed at any point(level) it is GAME OVER.
When you failed a question at any point the game restart automatically.

```python
Score = None
def quiz1 ():
    print("\nQuestions 1")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    
    Ans1 = input("Enter your Answer (i.e A/B/C/D)")    
    match Ans1:
        case "A":
            print("Correct")
            Score = 100
            print (f"Score: {Score}pts    Level: 1")
        case "B":
            print("Incorrect")
            Score = 0
            print (f"Score: {Score}pts    Level: 1")
        case "C":
            print("Incorrect")
            Score = 0
            print (f"Score: {Score}pts    Level: 1")
        case "D":
            print("Incorrect")
            Score = 0
            print (f"Score: {Score}pts    Level: 1")
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz1 ()
    return Score

def quiz2 (Score):
    print("\nQuestions 2")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    Ans2 = input("Enter your Answer (i.e A/B/C/D)")
        
    match Ans2:
        case "A":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 2")
        case "B":
            print("Correct")
            Score += 100
            print (f"Score: {Score}pts    Level: 2")
        case "C":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 2")
        case "D":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 2")
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz2 (Score)
    return Score

def quiz3 (Score):
    print("\nQuestions 3")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    Ans3 = input("Enter your Answer (i.e A/B/C/D)")
        
    match Ans3:
        case "A":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "B":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "C":
            print("Correct")
            Score += 200
            print (f"Score: {Score}pts    Level: 3")
        case "D":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz3 (Score)
    return Score

def quiz4 (Score):
    print("\nQuestions 4")
    print("A: Option 1")
    print("B: Option 2")
    print("C: Option 3")
    print("D: Option 4")
    Ans3 = input("Enter your Answer (i.e A/B/C/D)")
        
    match Ans3:
        case "A":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "B":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "C":
            print("Incorrect")
            Score += 0
            print (f"Score: {Score}pts    Level: 3")
        case "D":
            print("Correct")
            Score += 400
            print (f"Score: {Score}pts    Level: 3")
        case _:
            print("\nInvalid Input")
            print("You have enter an Invalid Option")
            Score = quiz3 (Score)
    return Score

def quizGame():
    Score = quiz1()
    
    while Score == 0:
        print("\nGame Over ReStarting.........................")
        Score = quiz1()

    Score2 = quiz2(Score)
    
    if Score2 == 100:
        print("\nGame Over ReStarting.........................")
        final = quizGame()
   
    else:
        Score3 = quiz3(Score2)
        
        if Score3 == 200:
            print("\nGame Over ReStarting.........................")
            final = quizGame()
        
        else:
            final = quiz4(Score3)
            
            if final == 400:
                print("\nGame Over ReStarting.........................")
                final = quizGame()
        
            else:
                print("\nYour Total Score BreakDown")
                print(f"Level 1: {Score}")
                print(f"Level 2: {Score2}")
                print(f"Level 3: {Score3}")
                print(f"Level 4: {final}")
    
    return final
    

totalScore = quizGame()
print(f"Total Point: {totalScore}")
```
